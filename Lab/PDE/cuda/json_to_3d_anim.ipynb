```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import json
```

```python
with open("build/x.json", "r") as file:
    xdata = np.array(json.load(file))
    file.close()
with open("build/y.json", "r") as file:
    ydata = np.array(json.load(file))
    file.close()
with open("build/z.json", "r") as file:
    zdata = np.array(json.load(file))
    file.close()

M, J = xdata.shape
```

```python
fig = plt.figure()
ax = plt.subplot(projection="3d")
line, = ax.plot([],[],[], "ro")
xpoints = np.zeros(J, dtype=float)
ypoints = np.zeros(J, dtype=float)
zpoints = np.zeros(J, dtype=float)
def init():
    return line,

def update(t):
    for i in range(J):
        xpoints[i] = xdata[t][i]
        ypoints[i] = ydata[t][i]
        zpoints[i] = zdata[t][i]
    line.set_data_3d(xpoints, ypoints, zpoints)
    return line,

ani = FuncAnimation(fig, update, frames=range(J), init_func=init, blit=True)

plt.show()
#update(10)
```
